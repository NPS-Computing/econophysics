# Chapter 3: Random Walk Model

```python
## Random Walk 1D
#'A guy is drunk! he either goes one step forward or one step backward'.
# Where will he be in 100 steps?
import matplotlib.pyplot as plt
%matplotlib inline
import random
def walk1D(x,N,d):
    X=[] # store position
    T=[] # store step
    for t in range(N):
        x=x+random.choice([-1,1])*d
        T.append(t)
        X.append(x)
    return T,X
```

```python
walk1D(0,10,1)
```

```python
time,pos=walk1D(0,100,1)
plt.plot(time,pos)
```

```python
walk1D(0,1000,1)
```

```python
time,pos=walk1D(0,1000,1)
plt.plot(time,pos)
```

```python
time,pos=walk1D(0,10000,1)
plt.plot(time,pos)
```
